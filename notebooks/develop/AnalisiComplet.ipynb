
## Imports i declaració de mètodes
Es declaren les variables globals i els mètodes per al càlcul de la distancia entre el centre de la ciutat i les mostres. També es defineixen els mètodes encarregats d'executar la regressió lineal.




```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import math
from sklearn import metrics
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
from math import radians, cos, sin, asin, sqrt
from scipy.stats import shapiro
from sklearn import metrics

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import shapiro
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression

# Localització del centre d'Amsterdam
latitude = 52.379189
longitude = 4.899431

def mean_squared_error(y1, y2):
    # comprovem que y1 i y2 tenen la mateixa mida
    assert(len(y1) == len(y2))
    mse = np.sum((y1-y2)**2)
    return mse / len(y1)

def regression(x, y):
    # Creem un objecte de regressió de sklearn
    regr = LinearRegression()
    # Entrenem el model per a predir y a partir de x
    regr.fit(x, y)
    # Retornem el model entrenat
    return regr

def predict_with_train(x, y, test_size, plot_text_x, plot_text_y):

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size)

    model = regression(x_train, y_train)
    y_hat = model.predict(x_test)

    # Mostrem el resultat
    plt.figure()
    plt.plot(x_train, y_train, '-o', alpha=0.25)
    plt.plot(x_test, y_hat, 'r', alpha=0.25)
    plt.xlabel(plot_text_y)
    plt.ylabel(plot_text_x)

    print("MSE:", mean_squared_error(y_hat, y_test))
    print ("R^2:", r2_score(y_hat, y_test))

def distance(lat1, lat2, lon1, lon2):

    lon1 = radians(lon1)
    lon2 = radians(lon2)
    lat1 = radians(lat1)
    lat2 = radians(lat2)

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2

    c = 2 * asin(sqrt(a))
    r = 6371

    # calculate the result
    return (c * r)

def dis(row):
    return distance(row['latitude'], latitude, row['longitude'], longitude)

def regression(x, y):
    # Creem un objecte de regressió de sklearn
    regr = LinearRegression()
    # Entrenem el model per a predir y a partir de x
    regr.fit(x, y)
    # Retornem el model entrenat
    return regr
```

## Adaptació del dataset
En aquest apartat generarem un atribut nou descartarem aquells que no siguin útils. També eliminem les mostres amb valors nulls i adaptem les dades que tinguin un format incorrecte.

L'atribut nou (distance_to_center) consisteix en la distància que hi ha entre el centre d'Amsterdam i l'allotjament en qüestió. Més endavant comprovarem si es tracta d'un factor determinant per la predicció del preu. 

```python
# Carreguem dataset d'exemple
dataset = pd.read_json("../../data/amsterdam.json")

# Visualitzem els nulls del dataset
null_columns = dataset.columns[dataset.isnull().any()]
print(null_columns)
print(dataset[null_columns].isnull().sum( ))
print("Total de valors no existents:", dataset.isnull().sum().sum())
# Eliminem les mostres amb nulls
dataset = dataset.dropna()
print("Total de valors no existents:", dataset.isnull().sum().sum())

data = dataset.values

# Creación y modificación de datos
dataset['distance_to_center'] = dataset.apply(lambda row: dis(row), axis=1)

le = LabelEncoder()
dataset['room_type'] = le.fit_transform(dataset['room_type'])
dataset['instant_bookable'] = le.fit_transform(dataset['instant_bookable'])

dataset['price'] = dataset['price'].str.replace('$', '')
dataset['price'] = dataset['price'].str.replace(',', '')
dataset['price'] = pd.to_numeric(dataset['price'])
```


## Regressió polinòmica

Repetim les proves anteriors, ara amb polinomis de diferents graus.

```python
def poli_with_train(degr, x, y, x_str, y_str, graph_name):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

    model = regression(x_train, y_train)
    y_hat = model.predict(x_test)

    x_train = x_train.values
    y_train = y_train.values

    x_test = x_test.values
    y_test = y_test.values

    x = x.values
    y = y.values

    poly_model = make_pipeline(PolynomialFeatures(degree=degr), linear_model.LinearRegression())
    poly_model.fit(x_train.reshape(-1, 1), y_train.reshape(-1, 1))

    linear_model_1 = linear_model.LinearRegression()
    linear_model_1.fit(x_train, y_train)

    fig = plt.figure()
    ax = plt.axes()

    x_test.sort(axis=0)

    ax.set(xlabel=x_str, ylabel=y_str, title=graph_name)
    ax.scatter(x,y, alpha=0.5)
    ax.plot(x_test, linear_model_1.predict(x_test), color='green', label='linear')
    ax.plot(x_test, poly_model.predict(x_test), color='red', label='poly')
    plt.show()

    print("MSE:", mean_squared_error(poly_model.predict(x_test),y_test))
    print("R^2:", r2_score(poly_model.predict(x_test), y_test))


y = dataset_editats[["price"]]

for index in range(2,10):
    print("------- degree = ", index, " -------")
    x = dataset_editats[['accommodates']]
    poli_with_train(index, x, y, 'accommodates', 'price', 'accommodates vs price')
    
for index in range(2,10):
    print("------- degree = ", index, " -------")
    x = dataset_editats[['bedrooms']]
    poli_with_train(index, x, y, 'bedrooms', 'price', 'bedrooms vs price')

```

Podem veure que els resultats no mostren millores respecte els resultats donats pel model de regressió lineal simple.
